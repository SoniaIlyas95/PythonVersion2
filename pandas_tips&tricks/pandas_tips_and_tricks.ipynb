# 02- Make a Dataframe
Whenever you work on dataset you need to create a datafram.

- To store the dataframe put them equall to any variable such as "df"
- To show the dataframe just put the command "df"
- You can change the length of column to work accordingly.

**__Note__: All Array must be of same length*


```python
df= pd.DataFrame({'A col': [1,2,3], 'B col':[4,5,6]})
df
```

## Numpy array dataframe
### 1st Method
- A dataframe can be created by using numpy array.
- numpy array can be converted into dataframe as shown:

```python
import numpy as np
arr = np.array([[1,2,3], [4,5,6], [3,6,4]])
pd.DataFrame(arr)
```

### 2nd method
Make a random numpy array as follows:

```python
pd.DataFrame(np.random.randn(4,8))
```

Numerical column can be changed into alphabetical signs.

*__Note__: Column list should be the same as the number of columns in datasets*

```python
pd.DataFrame(np.random.randn(4,8), columns=list('ABCDEFGH'))
```

# 03- Rename Column
First look at the simple dataframe. Then move towards the ways to rename column.

```python
df= pd.DataFrame({'A col': [1,2,3], 'B col':[4,5,6]})
df
```

## Ways to rename column
1- Apply df.rename and replace old name with new one.


```python
df.rename(columns={'A col':'A','B col':'B'}, inplace=True)
df
```

2- You can also change by just putting df.columns and their name(new names of columns)

```python
df.columns=['col_a','col_b']
df
```

3- To replace any character, string in dataframe column

```python
df.columns=df.columns.str.replace('_',' ')
df
```

## Adding Prefix And Suffix
__Prefix__ Means to add a constant before the name of each column in dataframe.

__Suffix__ Means to add a constant after the name of each column in dataframe.

```python
# Adding prefix
df = df.add_prefix('S_')
df
```

```python
# Adding Sufix
df = df.add_suffix('_ADD')
df
```

- If you want to return to their original column name simply rename them again by their real name.

```python
df.columns= ['col_a', 'col_b']
df
```
