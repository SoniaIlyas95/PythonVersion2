# 03- Rename Column
First look at the simple dataframe. Then move towards the ways to rename column.

```python
df= pd.DataFrame({'A col': [1,2,3], 'B col':[4,5,6]})
df
```

## Ways to rename column
1- Apply df.rename and replace old name with new one.


```python
df.rename(columns={'A col':'A','B col':'B'}, inplace=True)
df
```

2- You can also change by just putting df.columns and their name(new names of columns)

```python
df.columns=['col_a','col_b']
df
```

3- To replace any character, string in dataframe column

```python
df.columns=df.columns.str.replace('_',' ')
df
```

## Adding Prefix And Suffix
__Prefix__ Means to add a constant before the name of each column in dataframe.

__Suffix__ Means to add a constant after the name of each column in dataframe.

```python
# Adding prefix
df = df.add_prefix('S_')
df
```

```python
# Adding Sufix
df = df.add_suffix('_ADD')
df
```

- If you want to return to their original column name simply rename them again by their real name.

```python
df.columns= ['col_a', 'col_b']
df
```

# 14- Filtering a Datasets
After looking into the datasets find unique values in the datasets. 
- To check the unique values from the whole dataset use the code 'df.nunique()'
- you can also check the unique values of a particular column such as sex. you can use 'df.sex.unique()' code.
- If you want the data of only one category i.e male or female you can do it as 'df[(df.sex=="female")]' this will give you the column which contains only female.
- Similar to this you acn also sort other variables such as class by following same procedure.
- You can also specify/filter unique values of more than one variable such as embark_town and female at one time by using | and & function.
- 'isin' function is also used to filter the variable.

```python
df.head()
```

```python
df.sex.unique()
```

```python
df[(df.sex=="female")]
```

```python
df.nunique()
```

```python
df.alive.unique()
```

```python
print(df.embark_town.unique())
df[(df.embark_town=='Southampton')].shape
```

```python
df[((df.embark_town=='Southampton')|
    (df.embark_town=='Queenstown')) & 
    (df.sex=="female")]
```

```python
df[df.embark_town.isin(['Queenstown', 'Southampton'])].head()
```

```python
df[df.age < 18].shape
```

```python
df[((df.embark_town=='Southampton')|
    (df.embark_town=='Queenstown')) & 
    (df.sex=="female") & 
    (df.age > 45)].shape
```

# 15- Filtering by large categories
For datasets having large categories you can find the value count.
Similar to find the unique value of a particular variable you can actually find out the value count also. 
- A value_count give the uninque number as well as their values.
- If you desire to take only some largets 2 or 3 values then use this code 'df.age.value_counts().nlargest(3)'
- To include the index number in just follow 'df.age.value_counts().nlargest(3).index'
- 'nlargest' will give highest top class number.


```python
df.sex.value_counts()
```

```python
df.age.value_counts()
```

```python
df.age.value_counts().nlargest(3)
```

```python
df.age.value_counts().nlargest(3).index
```

```python
counts = df.who.value_counts()
counts.nlargest(3)
```

```python
df[df.who.isin(counts.nlargest(1).index)].head()
```
