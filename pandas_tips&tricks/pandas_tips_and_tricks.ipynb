# 03- Rename Column
First look at the simple dataframe. Then move towards the ways to rename column.

```python
df= pd.DataFrame({'A col': [1,2,3], 'B col':[4,5,6]})
df
```

## Ways to rename column
1- Apply df.rename and replace old name with new one.


```python
df.rename(columns={'A col':'A','B col':'B'}, inplace=True)
df
```

2- You can also change by just putting df.columns and their name(new names of columns)

```python
df.columns=['col_a','col_b']
df
```

3- To replace any character, string in dataframe column

```python
df.columns=df.columns.str.replace('_',' ')
df
```

## Adding Prefix And Suffix
__Prefix__ Means to add a constant before the name of each column in dataframe.

__Suffix__ Means to add a constant after the name of each column in dataframe.

```python
# Adding prefix
df = df.add_prefix('S_')
df
```

```python
# Adding Sufix
df = df.add_suffix('_ADD')
df
```

- If you want to return to their original column name simply rename them again by their real name.

```python
df.columns= ['col_a', 'col_b']
df
```

# 06- Reverse Row Order
In order to reverse the order of rows you need to import libraries and load dataset. You can also use your own data to reverse row order.
- First look at the dataset.
- Use the formula of reverse row order.

```python
import pandas as pd
import seaborn as sns

df = sns.load_dataset("titanic")
df.head(-6)
```

```python
df.loc[::-1].head(-5)
```

- Here indexes are also in reverse order. if you want to start it from zero. you need to reset index.

```python
df.loc[::-1].reset_index(drop=True).head()
```

# 07- Reverse Column Order
Similar to the rows columns can also be reversed by following code.

```python
df.loc[:, ::-1].head()
```

# 08- Select a Column by Datatype(dtype)
Based on datatype one can select particular column. just use 'df.dtype' code and type of every column e.g survived, pclass, sex, age, etc. will be shown in output.

```python
df.dtypes
```

- If you want to select only those have numeric type, put the code shown.
All the column having numeric values will appear here.

```python
df.select_dtypes(include=['number']).head()
```

- Similar to numeric if you want to select only those having object type, use the given code. 
All the columns with object type will appear here.

```python
df.select_dtypes(include=['object']).head()
```

- You can also select more than one type such as 'numeric' and 'categorical' at one time or 'object' and 'numericl' at one time in a dataframe or use combination of 'object', 'category' and 'number' as shown in code.

```python
df.select_dtypes(include=['object', 'category', 'number']).head()
```

- Just as the include function to represent in the dataframe you can also use exclude to remove a particular column from the dataframe. The simplest code is shown here.

```python
df.select_dtypes(exclude=['object']).head()
```

# 09- Convert String to Number
First you have to see the dtypes. Then convert the type in any int or float number by simply using the given code.

```python
df = pd.DataFrame({'col_A': ['1.2', '2', '3','6','7','8','9','22'], 
                    'col_B': ['4', '5', '6','7','8','9','10','11']})
df
```

```python
df.dtypes
```

```python
df = df.astype({'col_A': 'float64', 'col_B': 'int64'})
df
```

```python
df.dtypes
```

# 10- Reduce dataframe size
For very large datasets having thousands of rows, it will create difficulty in even loading the datasets. So, you need to reduce the size of dataframe.
- Different ways to reduce size:

1- See it according to your memory usage.

2- Reduce size by fraction value such as .1, .2, .5 etc. as in following codes.

```python
df = sns.load_dataset('titanic')
df.shape
```

```python
df.memory_usage(deep=True)
```

```python
df.sample(frac=0.1).shape
df.info()
```
